```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import tree   
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier 
from sklearn import metrics 
```

```python
data=pd.read_csv("titanic.csv")
data
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
x=data.drop(["Survived", "Ticket", "Embarked","Title"], axis=1)
y= data["Survived"]
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20,random_state=10)
dtree =DecisionTreeClassifier()  
dtree.fit(x_train, y_train)   
y_pred=dtree.predict(x_test)   
print("score:{}".format(accuracy_score(y_test, y_pred)))
```

```python
plt.figure(figsize=(30,20))
plot_tree(dtree,filled=True,feature_names = x.columns,proportion=True)
```

```python
x=data.drop(["Survived", "Ticket", "Embarked","Title"], axis=1)
y= data["Survived"]
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20,random_state=10)
dtree =DecisionTreeClassifier(max_leaf_nodes=15,max_depth=15)  
dtree.fit(x_train, y_train)   
y_pred=dtree.predict(x_test)   
print("score:{}".format(accuracy_score(y_test, y_pred)))
```

```python
plt.figure(figsize=(30,20))
plot_tree(dtree,filled=True,feature_names = x.columns,proportion=True)
```

```python

X=data[['Age', 'Sex', 'SibSp', 'Parch','Pclass','Fare']] 
y=data['Survived'] 
x_train, x_test, y_train, y_test= train_test_split(X, y, test_size=0.25) 
clf=RandomForestClassifier(n_estimators=50,max_leaf_nodes=15,max_depth=20)  
clf.fit(x_train, y_train)  
y_pred=clf.predict(x_test)  
print("Accuracy:", metrics.accuracy_score(y_test, y_pred)) 
```
